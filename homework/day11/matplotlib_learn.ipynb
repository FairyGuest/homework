```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']  # 用黑体来显示中文
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
x = np.linspace(0, 10, 100)
y = np.sin(x)  # sin函数

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='red', linewidth=2)
plt.title('正弦函数')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = 'upper right')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']  # 用黑体来显示中文
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.5, color='red', linewidth=2)
plt.title('散点图')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']  # 用黑体来显示中文
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
x = ['A', 'B', 'C', 'D', 'E']
y = [23, 13.4 , 56, 77.9 , 83]

plt.figure(figsize=(10, 6))
plt.bar(x, y, color='green')
plt.title('柱状图')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 用黑体来显示中文
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
sizes = [35, 25, 20, 15, 5]
labels = ['A', 'B', 'C', 'D', 'E']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']

plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # 确保饼图是圆的
plt.title('类别分布')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 用黑体来显示中文
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.figure(figsize=(10, 6))
plt.boxplot(data, labels=['A', 'B', 'C'])
plt.title('箱线图')
plt.ylabel('值')
plt.show()
```
